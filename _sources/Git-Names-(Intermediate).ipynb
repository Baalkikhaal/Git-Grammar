## HEAD

<span style="color: DodgerBlue">HEAD</span>
 is the points to the tip (latest commit) of the current branch.

A new <span style="color: DodgerBlue">commit</span>
 to the branch shifts the <span style="color: DodgerBlue">HEAD</span> to this new commit.
 
#### Notes

To reference to the penultimate commit the pointer name is <span style="color: DodgerBlue">HEAD^</span>. To reference the previous-previous commit the pointer name is <span style="color: DodgerBlue">HEAD~2</span>.

## SHA1

<span style="color: DodgerBlue">SHA1</span> is the algortithm used to <span style="color: DarkOrange">hash</span> the commits. This is the default hashing algorithm. Newer versions of git support <span style="color: DodgerBlue">SHA256</span> hashing algorithm.

## Hash

Hash is the process to create an encryption key for a blob. Hashing is a general computational process to provide keys that uniquely identify with a sequence of the bits involved in the blob. The hash key generated by git uses SHA-1 hashing algorithm. If we change even a single bit in the sequence, it leads to a completely different key.
